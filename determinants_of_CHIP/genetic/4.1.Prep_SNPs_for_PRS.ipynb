```r
## GWAS SNPs from REGENERON GWAS
ch_rg22_cojo <- read_excel("~/Documents/Project/Baylor_ARIC_Exomes/Paper1_ARIC/GWAS/41586_2022_5448_MOESM5_ESM.xlsx", 
                      sheet = 4)

ch_rg22_cojo$REF_Allele <- str_split_fixed(string = ch_rg22_cojo$SNP, 
                                    pattern = "[:]", n = 4)[,3]
ch_rg22_cojo$ALT_Allele <- str_split_fixed(string = ch_rg22_cojo$SNP, 
                                           pattern = "[:]", n = 4)[,4]
  # only SNPs with P<5e-8 and AF>5%
ch_rg22_cojo <- subset(ch_rg22_cojo, 
                       ch_rg22_cojo$P_COJO<5e-8 & 
                         ch_rg22_cojo$AAF>=0.05 &
                         ch_rg22_cojo$REF_Allele %in% 
                         c("A", "C", "G", "T") & 
                         ch_rg22_cojo$ALT_Allele %in% 
                         c("A", "C", "G", "T"))


  # DNMT3A
dnmt3a_rg22_cojo <- read_excel("~/Documents/Project/Baylor_ARIC_Exomes/Paper1_ARIC/GWAS/41586_2022_5448_MOESM5_ESM.xlsx", 
                          sheet = 6)

dnmt3a_rg22_cojo$REF_Allele <- str_split_fixed(string = dnmt3a_rg22_cojo$SNP, 
                                           pattern = "[:]", n = 4)[,3]
dnmt3a_rg22_cojo$ALT_Allele <- str_split_fixed(string = dnmt3a_rg22_cojo$SNP, 
                                           pattern = "[:]", n = 4)[,4]
# only SNPs with P<5e-8 and AF>5%
dnmt3a_rg22_cojo <- subset(dnmt3a_rg22_cojo, 
                           dnmt3a_rg22_cojo$P_COJO<5e-8 & 
                             dnmt3a_rg22_cojo$AAF>=0.05 &
                             dnmt3a_rg22_cojo$REF_Allele %in% 
                         c("A", "C", "G", "T") & 
                           dnmt3a_rg22_cojo$ALT_Allele %in% 
                         c("A", "C", "G", "T"))
```

```r

## lift0ver hg38 to hg19
library(liftOver)
library(rtracklayer)
library(TxDb.Hsapiens.UCSC.hg38.knownGene)
library(GenomicRanges)
# Load Chain file
hg38_to_19_chain = import.chain( "~/Documents/Project/Baylor_ARIC_Exomes/Paper1_ARIC/GWAS/hg38ToHg19.over.chain")

# CH
ch_rg22_cojo$CHROM <- paste0("chr",ch_rg22_cojo$CHR)
ch_rg22_cojo_df <- ch_rg22_cojo[,c(19, 3, 3,1)]
names(ch_rg22_cojo_df) <- c("chrom", "start", "end","varID")
gr.ch.hg38 <- as(ch_rg22_cojo_df, "GenomicRanges")

ch_hg38tohg19 <- liftOver(gr.ch.hg38, hg38_to_19_chain)
ch_hg38tohg19 <- as.data.frame(ch_hg38tohg19)

ch_hg38tohg19$chr_hg19 <- gsub(pattern = "chr",replacement = "", 
                                   x = as.character(ch_hg38tohg19$seqnames))

ch_hg38tohg19$POS_hg19 <- ch_hg38tohg19$start

ch_hg19_cojo <- merge(ch_hg38tohg19[,c(8,9,10)], 
                       ch_rg22_cojo, 
                           by.x="varID",
                       by.y="SNP")

ch_hg19_cojo$varID_hg19 <- paste(ch_hg19_cojo$chr_hg19, 
                                 ch_hg19_cojo$POS_hg19, 
                                 ch_hg19_cojo$REF_Allele, 
                                 ch_hg19_cojo$ALT_Allele, 
                                 sep=":")

table(ch_hg19_cojo$varID_hg19 %in% aric_afr_snp$varID )
# ch_hg19_cojo$varID_hg19[!(ch_hg19_cojo$varID_hg19 %in% aric_afr_snp$varID)]
table(ch_hg19_cojo$varID_hg19 %in% aric_eur_snp$varID )
# ch_hg19_cojo$varID_hg19[!(ch_hg19_cojo$varID_hg19 %in% aric_eur_snp$varID)]
# 21 SNPs
aa_aric_snvs_ch_hg19_cojo <- merge(ch_hg19_cojo[,c(1,2,3,19,20,
                                                           15,22,17,10)],
                                       aric_afr_snp, 
                                       by.x="varID_hg19",
                                       by.y="varID")

ea_aric_snvs_ch_hg19_cojo <- merge(ch_hg19_cojo[,c(1,2,3,19,20,
                                                           15,22,17,10)],
                                       aric_eur_snp, 
                                       by.x="varID_hg19",
                                       by.y="varID")
write.csv(ea_aric_snvs_ch_hg19_cojo, 
          "/Volumes/medpop_esp2/mesbah/projects/ch_progression/aric/gwas/ea_aric_21snvs_ch_hg19_cojo.csv", 
          row.names = F)
write.csv(aa_aric_snvs_ch_hg19_cojo, 
          "/Volumes/medpop_esp2/mesbah/projects/ch_progression/aric/gwas/aa_aric_21snvs_ch_hg19_cojo.csv", 
          row.names = F)
## DNMT3A
dnmt3a_rg22_cojo$CHROM <- paste0("chr",dnmt3a_rg22_cojo$CHR)
dnmt3a_rg22_cojo_df <- dnmt3a_rg22_cojo[,c(19, 3, 3,1)]
names(dnmt3a_rg22_cojo_df) <- c("chrom", "start", "end","varID")
gr.dnmt3a.hg38 <- as(dnmt3a_rg22_cojo_df, "GenomicRanges")

dnmt3a_hg38tohg19 <- liftOver(gr.dnmt3a.hg38, 
                              hg38_to_19_chain)
dnmt3a_hg38tohg19 <- as.data.frame(dnmt3a_hg38tohg19)

dnmt3a_hg38tohg19$chr_hg19 <- gsub(pattern = "chr",replacement = "", 
                               x = as.character(dnmt3a_hg38tohg19$seqnames))

dnmt3a_hg38tohg19$POS_hg19 <- dnmt3a_hg38tohg19$start

dnmt3a_hg19_cojo <- merge(dnmt3a_hg38tohg19[,c(8,9,10)], 
                          dnmt3a_rg22_cojo, 
                          by.x="varID",
                          by.y="SNP")

dnmt3a_hg19_cojo$varID_hg19 <- paste(dnmt3a_hg19_cojo$chr_hg19, 
                                     dnmt3a_hg19_cojo$POS_hg19, 
                                     dnmt3a_hg19_cojo$REF_Allele, 
                                     dnmt3a_hg19_cojo$ALT_Allele, 
                                 sep=":")



# 22 SNPs
table(dnmt3a_hg19_cojo$varID_hg19 %in% aric_afr_snp$varID )

table(dnmt3a_hg19_cojo$varID_hg19 %in% aric_eur_snp$varID )

aa_aric_snvs_dnmt3a_hg19_cojo <- merge(dnmt3a_hg19_cojo[,c(1,2,3,19,20,
                                                           15,22,17,10)],
                                     aric_afr_snp, 
                                     by.x="varID_hg19",
                                     by.y="varID")

ea_aric_snvs_dnmt3a_hg19_cojo <- merge(dnmt3a_hg19_cojo[,c(1,2,3,19,20,
                                                         15,22,17,10)],
                                     aric_eur_snp, 
                                     by.x="varID_hg19",
                                     by.y="varID")
write.csv(ea_aric_snvs_dnmt3a_hg19_cojo, 
          "/Volumes/medpop_esp2/mesbah/projects/ch_progression/aric/gwas/ea_aric_22snvs_dnmt3a_hg19_cojo.csv", 
          row.names = F)
write.csv(aa_aric_snvs_dnmt3a_hg19_cojo, 
          "/Volumes/medpop_esp2/mesbah/projects/ch_progression/aric/gwas/aa_aric_22snvs_dnmt3a_hg19_cojo.csv", 
          row.names = F)
## TET2
## TET2
tet2_rg22_indx <- read_excel("~/Documents/Project/Baylor_ARIC_Exomes/Paper1_ARIC/GWAS/41586_2022_5448_MOESM5_ESM.xlsx", 
                             sheet = 15)

tet2_rg22_indx$REF_Allele <- str_split_fixed(string = tet2_rg22_indx$name, 
                                             pattern = "[:]", n = 4)[,3]
tet2_rg22_indx$ALT_Allele <- str_split_fixed(string = tet2_rg22_indx$name, 
                                             pattern = "[:]", n = 4)[,4]

tet2_rg22_indx$Beta <- log(tet2_rg22_indx$effect_UKB)

# only SNPs with P<5e-8 and AF>5%
tet2_rg22_indx <- subset(tet2_rg22_indx, 
                         tet2_rg22_indx$pval_UKB<5e-8 & 
                           tet2_rg22_indx$aaf_UKB>=0.05 &
                           tet2_rg22_indx$REF_Allele %in% 
                           c("A", "C", "G", "T") & 
                           tet2_rg22_indx$ALT_Allele %in% 
                           c("A", "C", "G", "T"))

tet2_rg22_indx$CHROM <- paste0("chr",tet2_rg22_indx$chr)

tet2_rg22_cojo_df <- tet2_rg22_indx[,c(45, 3, 3,1)]
names(tet2_rg22_cojo_df) <- c("chrom", "start", "end","varID")
gr.tet2.hg38 <- as(tet2_rg22_cojo_df, "GenomicRanges")

tet2_hg38tohg19 <- liftOver(gr.tet2.hg38, 
                              hg38_to_19_chain)

tet2_hg38tohg19 <- as.data.frame(tet2_hg38tohg19)

tet2_hg38tohg19$chr_hg19 <- gsub(pattern = "chr",replacement = "", 
                                   x = as.character(tet2_hg38tohg19$seqnames))

tet2_hg38tohg19$POS_hg19 <- tet2_hg38tohg19$start

tet2_hg19_cojo <- merge(tet2_hg38tohg19[,c(8,9,10)], 
                          tet2_rg22_indx, 
                          by.x="varID",
                          by.y="name")

tet2_hg19_cojo$varID_hg19 <- paste(tet2_hg19_cojo$chr_hg19, 
                                   tet2_hg19_cojo$POS_hg19, 
                                   tet2_hg19_cojo$REF_Allele, 
                                   tet2_hg19_cojo$ALT_Allele, 
                                     sep=":")
# 6 SNPs
table(tet2_hg19_cojo$varID_hg19 %in% aric_afr_snp$varID )

table(tet2_hg19_cojo$varID_hg19 %in% aric_eur_snp$varID )

aa_aric_snvs_tet2_hg19_cojo <- merge(tet2_hg19_cojo[,c(1,2,3,44,45,46,48,13,14)],
                                     aric_afr_snp, 
                                     by.x="varID_hg19",
                                     by.y="varID")

ea_aric_snvs_tet2_hg19_cojo <- merge(tet2_hg19_cojo[,c(1,2,3,44,45,46,48,13,14)],
                                     aric_eur_snp, 
                                     by.x="varID_hg19",
                                     by.y="varID" 
                                     )
write.csv(ea_aric_snvs_tet2_hg19_cojo, 
          "/Volumes/medpop_esp2/mesbah/projects/ch_progression/aric/gwas/ea_aric_6snvs_tet2_hg19_cojo.csv", 
          row.names = F)
write.csv(aa_aric_snvs_tet2_hg19_cojo, 
          "/Volumes/medpop_esp2/mesbah/projects/ch_progression/aric/gwas/aa_aric_6snvs_tet2_hg19_cojo.csv", 
          row.names = F)
```
